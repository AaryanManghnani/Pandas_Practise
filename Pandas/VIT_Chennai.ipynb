```python
import PyPDF2

def read_pdf_and_concatenate_text(pdf_path):

  pdf_reader = PyPDF2.PdfReader(pdf_path)
  concatenated_text = "".join(page.extract_text() for page in pdf_reader.pages)
  return concatenated_text

path = "C:/College Stuff/CV/Aaryan Manghnani CV 2.0.pdf"
string = read_pdf_and_concatenate_text(path)
print(string)
```
